#### Create a boolean mask by iterating through the entire dataframe and use the boolean mask

```python
import pandas as pd
```

```python
df=pd.read_csv('../Datasets/nba-2.csv')
```

```python
df.head(10)
```

## Using boolean mask:

### Finding all players below the age of 25 having salary greater than 1000000.0

```python
temp=[]

for i in range(len(df)):
    if((df.loc[i,'Age']< 25.0) and (df.loc[i,'Salary']>1000000.0)):
        temp.append(True)
    else:
        temp.append(False)
```

```python
df[temp]
```
